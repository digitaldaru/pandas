## 11. How to Create a Dataframe

```python
import pandas as pd 

lst = [['apple', 'red', 11],
       ['grape', 'green', 22], 
       ['orange', 'orange', 33], 
       ['mango', 'yellow', 44]] 
df = pd.DataFrame(lst, columns =['Fruits', 'Color', 'Value']) 
print(df) 
```

```python
n = ["apple", "grape", "orange", "mango"] 
col = ["red", "green", "orange", "yellow"] 
val = [11, 22, 33, 44] 

dict = {'fruit': n, 'color': col, 'value': val}  
df = pd.DataFrame(dict) 
print(df) 
```

```python
df.to_csv("Datasets/fruits.csv",index=False)
```
